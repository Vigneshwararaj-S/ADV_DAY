```python
def split_text(text, max_length=500):
    print(text)
    """Splits text into smaller parts with a maximum length."""
    return [text[i:i+max_length] for i in range(0, len(text), max_length)]

def send_in_batches(documents, batch_size=5):
    """Function to send documents in batches of 5."""
    for i in range(0, len(documents), batch_size):
        yield documents[i:i+batch_size]

# Split the text in the 'documents' variable into smaller parts
split_documents = [split_text(doc) for doc in documents]
print(split_documents)
print(len(split_documents))

# Flatten the list of lists
split_documents = [item for sublist in split_documents for item in sublist]
print(split_documents)

# Send the split documents to the text analytics client in batches
try:
    for batch in send_in_batches(split_documents):
        response = text_analytics_client.recognize_entities(batch)
        for idx, phrase in enumerate(response):
            for entity in phrase.entities:
                if entity.category == "Email":
                    print("Entity: ", entity.text, "Type: ", entity.category)
except Exception as e:
    print("Error:", e)


```

```python
def extract_key_phrases(log_list):
    response = text_analytics_client.extract_key_phrases(documents=log_list)
    key_phrases = []
    for doc in response:
        if not doc.is_error:
            key_phrases.append(",".join(doc.key_phrases))
        else:
            key_phrases.append("Error extracting key phrases")
    return key_phrases
```

```python
from sklearn.metrics import silhouette_score

def determine_optimal_clusters_silhouette(X, max_clusters=10):
    best_n_clusters = 2
    best_score = -1

    for i in range(2, max_clusters + 1):  # Start from 2 because silhouette score requires at least 2 clusters
        kmeans = KMeans(n_clusters=i, random_state=42)
        cluster_labels = kmeans.fit_predict(X)
        silhouette_avg = silhouette_score(X, cluster_labels)
        
        print(f"For n_clusters = {i}, the silhouette score is {silhouette_avg:.4f}")
        
        if silhouette_avg > best_score:
            best_n_clusters = i
            best_score = silhouette_avg

    return best_n_clusters

# Determine optimal clusters using silhouette score
optimal_n_clusters = determine_optimal_clusters_silhouette(X, max_clusters=10)
kmeans = KMeans(n_clusters=optimal_n_clusters, random_state=42)
clusters = kmeans.fit_predict(X)
```

```python
def create_dynamic_root_cause(logs, key_phrases, cluster_assignments):
    cluster_root_causes = {}
    # Group logs by their cluster and key phrases
    cluster_logs = {}
    
    for i, cluster in enumerate(cluster_assignments):
        if cluster not in cluster_logs:
            cluster_logs[cluster] = []
        cluster_logs[cluster].append(key_phrases[i])
    
    # For each cluster, find the most frequent key phrases to generate a dynamic cause name
    for cluster, phrases in cluster_logs.items():
        all_phrases = " ".join(phrases).split()
        most_common_phrases = Counter(all_phrases).most_common(3)  # Top 2 common phrases
        cause_name = " ".join([phrase for phrase, _ in most_common_phrases])
        cluster_root_causes[cluster] = cause_name
    
    return cluster_root_causes
```

```python
dynamic_root_causes = create_dynamic_root_cause(logs, key_phrases, clusters)
```
